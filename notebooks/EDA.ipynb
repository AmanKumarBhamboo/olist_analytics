## 6. Order Reviews Dataset
Customer reviews with scores, optional comments, and timestamps.

```python
df_reviews = pd.read_csv('raw_data/olist_order_reviews_dataset.csv')
print('Shape:', df_reviews.shape)
print('\nInfo:')
df_reviews.info()
print('\nHead:')
df_reviews.head()
```

```python
print('Missing values:')
print(df_reviews.isnull().sum())
print('\nReview score distribution:')
print(df_reviews['review_score'].value_counts().sort_index())
```

```python
# Review comments are optional — fill missing text with empty string
df_reviews['review_comment_title'] = df_reviews['review_comment_title'].fillna('')
df_reviews['review_comment_message'] = df_reviews['review_comment_message'].fillna('')

# Drop rows without review_score (critical for analysis) and order_id
print(f'Dropping {df_reviews["review_score"].isna().sum()} rows with missing review_score')
df_reviews = df_reviews.dropna(subset=['review_score', 'order_id'])

# Fill other timestamps with a placeholder
df_reviews['review_creation_date'] = df_reviews['review_creation_date'].fillna('unknown')
df_reviews['review_answer_timestamp'] = df_reviews['review_answer_timestamp'].fillna('unknown')

print('Shape after handling:', df_reviews.shape)
print('\nRemaining missing values:')
print(df_reviews.isnull().sum())
```

## 7. Products Dataset
Product metadata: category, dimensions, weight, photo count.

```python
df_products = pd.read_csv('raw_data/olist_products_dataset.csv')
print('Shape:', df_products.shape)
print('\nInfo:')
df_products.info()
print('\nHead:')
df_products.head()
```

```python
print('Missing values:')
print(df_products.isnull().sum())
print('\nTop product categories:')
print(df_products['product_category_name'].value_counts().head(15))
print('\nWeight stats:')
print(df_products['product_weight_g'].describe())
```

```python
drop_cols = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty',
            'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
df_products = df_products.drop(columns=drop_cols)
print('Shape after drop:', df_products.shape)
print('\nRemaining columns:', list(df_products.columns))
df_products.head()
```

```python
# ~610 products have missing category name (and related text/photo columns)
# Fill category with 'unknown'
print(f'Products with missing category: {df_products["product_category_name"].isna().sum()}')
df_products['product_category_name'] = df_products['product_category_name'].fillna('unknown')

# 2 products have missing weight/dimensions — fill with median
for col in ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']:
    if col in df_products.columns:
        median_val = df_products[col].median()
        print(f'Filling {col} with median: {median_val}')
        df_products[col] = df_products[col].fillna(median_val)

print('\nMissing values after handling:')
print(df_products.isnull().sum())
```

## Handling the duplicated values

```python
# Geolocation: 261,831 fully duplicate rows — drop them
print(f'Before: {df_geolocation.shape}')
df_geolocation = df_geolocation.drop_duplicates()
print(f'After: {df_geolocation.shape}')
```

```python
# Reviews: 789 duplicate review_ids — keep first occurrence
print(f'Before: {df_reviews.shape}')
print(f'Duplicate review_ids: {df_reviews["review_id"].duplicated().sum()}')
df_reviews = df_reviews.drop_duplicates(subset=['review_id'], keep='first')
print(f'After: {df_reviews.shape}')
```

```python
# Map state abbreviations to full names (customers & geolocation)state_name_map = {    'AC': 'Acre', 'AL': 'Alagoas', 'AM': 'Amazonas', 'AP': 'Amapá',    'BA': 'Bahia', 'CE': 'Ceará', 'DF': 'Distrito Federal', 'ES': 'Espírito Santo',    'GO': 'Goiás', 'MA': 'Maranhão', 'MG': 'Minas Gerais', 'MS': 'Mato Grosso do Sul',    'MT': 'Mato Grosso', 'PA': 'Pará', 'PB': 'Paraíba', 'PE': 'Pernambuco',    'PI': 'Piauí', 'PR': 'Paraná', 'RJ': 'Rio de Janeiro', 'RN': 'Rio Grande do Norte',    'RO': 'Rondônia', 'RR': 'Roraima', 'RS': 'Rio Grande do Sul', 'SC': 'Santa Catarina',    'SE': 'Sergipe', 'SP': 'São Paulo', 'TO': 'Tocantins',}df_customers['customer_state'] = df_customers['customer_state'].map(state_name_map).fillna(df_customers['customer_state'])df_geolocation['geolocation_state'] = df_geolocation['geolocation_state'].map(state_name_map).fillna(df_geolocation['geolocation_state'])print(f'Unique customer states: {df_customers["customer_state"].nunique()}')print(f'Unique geolocation states: {df_geolocation["geolocation_state"].nunique()}')
```

```python
from sqlalchemy import create_engine

engine = create_engine('postgresql://apple@localhost:5432/olist')

tables = {
    'customers': df_customers,
    'geolocation': df_geolocation,
    'orders': df_orders,
    'order_items': df_order_items,
    'order_payments': df_payments,
    'order_reviews': df_reviews,
    'products': df_products,
    'sellers': df_sellers,
    'product_category_translation': df_category_translation,
}

for name, df in tables.items():
    print(f'Exporting {name} ({len(df)} rows)...', end=' ')
    df.to_sql(name, engine, if_exists='replace', index=False)
    print('done')

print('\nAll tables exported successfully.')
```
